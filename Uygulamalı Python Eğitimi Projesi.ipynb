```python
import os



GOREV_DOSYASI = "gorevler.txt"


def gorevleri_yukle():
    try:
        with open(GOREV_DOSYASI, "r", encoding="utf-8") as dosya:
            gorevler = [satir.strip() for satir in dosya.readlines()]
        return gorevler
    except FileNotFoundError:
        return []
    except Exception as e:
        print(f"Hata oluştu: {e}")
        return []
```

```python
def gorevleri_kaydet(gorevler):
    try:
        with open(GOREV_DOSYASI, "w", encoding="utf-8") as dosya:
            for gorev in gorevler:
                dosya.write(gorev + "\n")
    except Exception as e:
        print(f"Hata oluştu: {e}")
```

```python
def gorevleri_listele(gorevler):
    if not gorevler:
        print("Henüz eklenmiş bir görev yok.")
    else:
        print("\n--- Görev Listesi ---")
        for i, gorev in enumerate(gorevler, 1):
            print(f"{i}. {gorev}")
        print("---------------------\n")
```

```python

def yeni_gorev_ekle(gorevler):
    gorev = input("Yeni görevi girin: ").strip()
    if gorev:
        gorevler.append(gorev)
        gorevleri_kaydet(gorevler)
        print("Görev başarıyla eklendi.\n")
    else:
        print("Boş görev eklenemez!\n")
```

```python

def gorev_sil(gorevler):
    gorevleri_listele(gorevler)
    try:
        secim = int(input("Silmek istediğiniz görevin numarasını girin: "))
        if 1 <= secim <= len(gorevler):
            silinen = gorevler.pop(secim - 1)
            gorevleri_kaydet(gorevler)
            print(f"'{silinen}' görevi silindi.\n")
        else:
            print("Geçersiz görev numarası!\n")
    except ValueError:
        print("Lütfen geçerli bir sayı girin!\n")
```

```python
def gorev_duzenle(gorevler):
    gorevleri_listele(gorevler)
    try:
        secim = int(input("Düzenlemek istediğiniz görevin numarasını girin: "))
        if 1 <= secim <= len(gorevler):
            yeni_gorev = input("Yeni görev metnini girin: ").strip()
            if yeni_gorev:
                eski = gorevler[secim - 1]
                gorevler[secim - 1] = yeni_gorev
                gorevleri_kaydet(gorevler)
                print(f"'{eski}' görevi '{yeni_gorev}' olarak güncellendi.\n")
            else:
                print("Boş görev metni girilemez!\n")
        else:
            print("Geçersiz görev numarası!\n")
    except ValueError:
        print("Lütfen geçerli bir sayı girin!\n")
```

```python
def ana_menu():
    gorevler = gorevleri_yukle()
    while True:
        print("\n--- GÖREV YÖNETİCİSİ ---")
        print("1. Görevleri Listele")
        print("2. Yeni Görev Ekle")
        print("3. Görev Düzenle")
        print("4. Görev Sil")
        print("5. Çıkış")

        secim = input("Seçiminizi girin (1-5): ").strip()
        if secim == "1":
            gorevleri_listele(gorevler)
        elif secim == "2":
            yeni_gorev_ekle(gorevler)
        elif secim == "3":
            gorev_duzenle(gorevler)
        elif secim == "4":
            gorev_sil(gorevler)
        elif secim == "5":
            print("Çıkılıyor... bize vakit ayırdığınız için teşekkürler:)")
            break
        else:
            print("Geçersiz seçim! Lütfen 1 ile 5 arasında bir değer girin.\n")


if __name__ == "__main__":
    ana_menu()
```
